## Bar Chart Solution - Practice 1

## https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc

```python
import matplotlib.pyplot as plt
```

```python
# simple Bar chart - Top 10 Movie Box Office
top_10_imdb_movies = ["Shawshank Redemption", "The Godfather", "The Dark Knight" , "The Lord of the Rings III",
                      "Schindler's List","The Godfather II","12 Angry Men","The Lord of the Rings I",
                      "Pulp Fiction","Inception"]
worldwide_boxoffice_dollar = [29331551, 250342198, 1009025676, 1138027091, 322161245, 47961850, 2945, 888120042,
                             213928762, 839030630]
# set the plot size to increase the width
plt.rcParams["figure.figsize"] = (20,5)
ax = plt.bar(top_10_imdb_movies, worldwide_boxoffice_dollar)
plt.xlabel("movies")
plt.ylabel("box office in $")
plt.show()
```

## Bar Chart Solution - Practice 2

## https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc

```python
# histogram - IMDB Top 10 Movies Rating

ratings = [9.3, 9.2, 9.0, 9.0, 9.0, 9.0, 9.0, 8.9, 8.9, 8.8]
plt.hist(ratings)
plt.title("Distribution of top 10 movies ratings")
plt.xlabel("range of ratings")
plt.ylabel("number of movies")
plt.grid()
plt.show()
```
