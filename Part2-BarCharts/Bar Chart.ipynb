```python
import matplotlib.pyplot as plt
```

```python
# simple Bar chart - people's age
names = ["Ali", "Armita", "Kia"]
age = ["32", "37", "29"]
plt.bar(names, age)
plt.show()
```

```python
# simple Bar chart - Movies Oscar winning
movies = ["Annie Hall", "Ben-Hur", "Casablanca" , "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

plt.bar(movies, num_oscars)
plt.xlabel("movies")
plt.ylabel("number of oscars")
plt.grid()
plt.show()
```

```python
# Bar Color
# simple Bar chart - Movies Oscar winning
movies = ["Annie Hall", "Ben-Hur", "Casablanca" , "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

plt.bar(movies, num_oscars, color="green")
plt.xlabel("movies")
plt.ylabel("number of oscars")
plt.grid()
plt.show()
```

```python
# add Width
# simple Bar chart - Movies Oscar winning
movies = ["Annie Hall", "Ben-Hur", "Casablanca" , "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

plt.bar(movies, num_oscars, color="green", width= 0.3)
plt.xlabel("movies")
plt.ylabel("number of oscars")
plt.grid()
plt.show()
```

```python
# Horizontal Bar chart
countries = ["Iran", "Turkey", "Germany", "Thailand"]
population = [89.173, 85.816, 83.294, 71.801 ]

plt.barh(countries, population)
plt.show()
```

```python
# add Height 
countries = ["Iran", "Turkey", "Germany", "Thailand"]
population = [89.173, 85.816, 83.294, 71.801 ]

plt.barh(countries, population, height = 0.3)
plt.show()
```

```python
# histogram - Exam grades

grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
plt.hist(grades)
plt.title("Distribution of Exam grades")
plt.xlabel("range of grades")
plt.ylabel("number of students")
plt.grid()
plt.show()
```
