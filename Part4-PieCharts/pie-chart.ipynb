```python
import matplotlib.pyplot as plt
```

```python
# pie chart 1 - daily routine tasks (24 hr)

routine = [8, 8, 2, 1, 3, 2]

plt.pie(routine)
plt.show()
```

```python
# pie chart 2 - countries population density per km2

density =[481, 152, 37, 153, 312]
plt.pie(density)
plt.show()
```

```python
# add labels
density =[481, 152, 37, 153, 312]
lbs = ["India", "China", "USA", "Indonesia", "Pakistan"]
plt.pie(density, labels= lbs)
plt.show()
```

```python
# add colors

density =[481, 152, 37, 153, 312]
lbs = ["India", "China", "USA", "Indonesia", "Pakistan"]
clr = ["green", "blue", "yellow", "red", "grey"]
plt.pie(density, labels= lbs, colors= clr)
plt.show()
```

```python
# start Angle

density =[481, 152, 37, 153, 312]
lbs = ["India", "China", "USA", "Indonesia", "Pakistan"]
clr = ["green", "blue", "yellow", "red", "grey"]
plt.pie(density, labels= lbs, colors= clr, startangle= 180)
plt.show()
```

```python
# add legend 

density =[481, 152, 37, 153, 312]
lbs = ["India", "China", "USA", "Indonesia", "Pakistan"]
clr = ["green", "blue", "yellow", "red", "grey"]
plt.pie(density, labels= lbs, colors= clr)
plt.legend()
plt.show()
```

```python
# add legend with header

density =[481, 152, 37, 153, 312]
lbs = ["India", "China", "USA", "Indonesia", "Pakistan"]
clr = ["green", "blue", "yellow", "red", "grey"]
plt.pie(density, labels= lbs, colors= clr)
plt.legend(title="countries")
plt.show()
```

```python
# add percentage into the chart

density =[481, 152, 37, 153, 312]
lbs = ["India", "China", "USA", "Indonesia", "Pakistan"]
clr = ["green", "blue", "yellow", "red", "grey"]
plt.pie(density, labels= lbs, colors= clr, autopct='%1.1f%%')
plt.show()
```
