## Line Chart Solution - Practice 1

## https://www.statista.com/statistics/188105/annual-gdp-of-the-united-states-since-1990/
## https://www.statista.com/statistics/281744/gdp-of-the-united-kingdom/

```python
import matplotlib.pyplot as plt
```

```python
# subplots - side by side
year = [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]

# the GDPs amount are in Billions
usa_gdp_billion = [16880.7, 17608.1, 18295, 18804.9, 19612.1, 20656.5, 21521.4, 21323, 23594, 25744.1, 27356.4]

# the GDP for UK were in Millions - We Converted them to Billions
uk_gdp_billion = [1963.8, 2026.5, 2071.5, 2111.4, 2167.4, 2197.8, 2233.9, 2002.5, 2176.2, 2270.8, 2273.1 ]

# plot 1 - left side
plt.subplot(1,3,1)
plt.plot(year, usa_gdp_billion, color='red', label='USA GDP 2013 - 2023', marker="o")
plt.title('USA GDP 2013 - 2023')

# plot 1 - right side
plt.subplot(1,3,3)
plt.plot(year, uk_gdp_billion, color='blue', label='UK GDP 2013 - 2023', marker="*")
plt.title('UK GDP 2013 - 2023')

plt.show()
```

```python
#plots - Multiline Charts

year = [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]

# the GDPs amount are in Billions
usa_gdp_billion = [16880.7, 17608.1, 18295, 18804.9, 19612.1, 20656.5, 21521.4, 21323, 23594, 25744.1, 27356.4]

# the GDP for UK were in Millions - We Converted them to Billions
uk_gdp_billion = [1963.8, 2026.5, 2071.5, 2111.4, 2167.4, 2197.8, 2233.9, 2002.5, 2176.2, 2270.8, 2273.1 ]

plt.plot(year, usa_gdp_billion, color='red', label='USA GDP 2013 - 2023', marker="o")
plt.plot(year, uk_gdp_billion, color='blue', label='UK GDP 2013 - 2023', marker="*")
plt.legend()
plt.title('USA and UK GDP 2013 - 2023')

plt.show()
```
