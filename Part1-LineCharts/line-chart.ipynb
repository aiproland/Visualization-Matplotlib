```python
import matplotlib.pyplot as plt
```

```python
# simple chart
data  = [12, 5, 6, 7, 8]
plt.plot(data)
plt.show()
```

```python
# add marker 
data  = [12, 5, 6, 7, 8]
plt.plot(data, marker="*")
plt.show()
```

```python
# show grid
# add marker 
data  = [12, 5, 6, 7, 8]
plt.plot(data, marker="o")
plt.grid()
plt.show()
```

```python
# add linestyle
data  = [12, 5, 6, 7, 8]
plt.plot(data, marker="o", linestyle="dashed")
#plt.grid()
plt.show()
```

```python
# add axis labels 
data  = [12, 5, 6, 7, 8]
plt.plot(data, marker="o", linestyle="dashed")
plt.xlabel("x label")
plt.ylabel("y label")
#plt.grid()
plt.show()
```

```python
# add plot title
data  = [12, 5, 6, 7, 8]
plt.plot(data, marker="o", linestyle="dashed")
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("this is title of the plot")
#plt.grid()
plt.show()
```

```python
# multi-line chart with legend

x1 = [45, 67, 89]
y1 = [15, 12, 20]
y2 = [14, 16, 28]
plt.plot(x1, y1, color='red', label='line one', marker="o")
plt.plot(x1, y2, color='blue', label='line two', marker="*")
plt.legend()
plt.show()
```

```python
# subplots - 2 columns
x1 = [45, 67, 89]
y1 = [15, 12, 20]
y2 = [14, 16, 28]

# plot 1 - left side
plt.subplot(1,2,1)
plt.plot(x1, y1, color='red', label='line one', marker="o")

# plot 1 - right side
plt.subplot(1,2,2)
plt.plot(x1, y2, color='blue', label='line two', marker="*")

plt.show()
```

```python
# subplots - 2 rows
x1 = [45, 67, 89]
y1 = [15, 12, 20]
y2 = [14, 16, 28]

# plot 1 - row1
plt.subplot(2,1,1)
plt.plot(x1, y1, color='red', label='line one', marker="o")

# plo2 - row2
plt.subplot(2,1,2)
plt.plot(x1, y2, color='blue', label='line two', marker="*")
plt.show()
```

```python
# exanple: gdp 
years = [1950, 1960, 1970, 1980, 1990, 2000]
gdp = [300.2, 543, 1075.5, 2045.6, 59000, 1490.4]

plt.plot(years, gdp, color="green", marker="o")
plt.title("GDP")
plt.xlabel("year")
plt.ylabel("gdp Milloins")

plt.show()
```

```python
# exanple: gdp 
days = ['Sat', 'Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri']
gdp = [26, 27, 27, 28, 30, 30, 32]

plt.plot(days, gdp, color="green", marker="o")
plt.title("Weather Temp in the last week")
plt.grid()
plt.xlabel("Weekdays")
plt.ylabel("Temprature C")

plt.show()
```
