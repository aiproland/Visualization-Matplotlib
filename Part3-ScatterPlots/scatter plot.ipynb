```python
import matplotlib.pyplot as plt
```

```python
# scatter plot - people's weight and height
weight = [50, 67, 85, 55, 70]
height = [165, 173, 155, 160, 180]

plt.scatter(weight, height)
plt.xlabel("weight")
plt.ylabel("height")
plt.show()
```

```python
# scatter plot - social website spend time
followers = [100, 6000, 575, 4000]
spend_time = [1, 5, 3, 6]

plt.scatter(followers, spend_time)
plt.show()
```

```python
# add color

followers = [100, 6000, 575, 4000]
spend_time = [1, 5, 3, 6]

plt.scatter(followers, spend_time, color="red")
plt.show()
```

```python
# add size

followers = [100, 6000, 575, 4000]
spend_time = [1, 5, 3, 6]
sizes = [40, 150, 65, 70]
plt.scatter(followers, spend_time, s=sizes)
plt.show()
```

```python
# Compare plots

m_weight = [50, 67, 85, 55, 70]
m_height = [165, 173, 155, 160, 180]

f_weight = [55, 60, 80, 53, 78]
f_height = [160, 175, 150, 165, 184]

# plot for men group
plt.scatter(m_weight, m_height, label = "men group")

# plot for females group
plt.scatter(f_weight, f_height, label = "females group")

plt.xlabel("weight")
plt.ylabel("height")
plt.legend()

plt.show()
```

```python
# add color map

m_weight = [50, 67, 85, 55, 70]
m_height = [165, 173, 155, 160, 180]
colors = [2, 35, 67, 78, 100]
# plot for men group
plt.scatter(m_weight, m_height, label = "men group", c=colors, cmap="viridis")

plt.colorbar()

plt.show()
```
